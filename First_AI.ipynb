```python
# Import libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from google.colab import drive
drive.mount('/content/drive')
# Path to answers CSV
affirmative = "/content/drive/MyDrive/affirmative.csv"
negative = "/content/drive/MyDrive/negative.csv"
# Function to determine answer type
def determine_answer(reply):
    reply = reply.lower()

    # Load affirmative + negative lists

    a = pd.read_csv(affirmative, header=None)[0].str.lower().tolist()
    n = pd.read_csv(negative, header=None)[0].str.lower().tolist()

    # Check if reply matches known answers
    if reply in a:
        return 1
    if reply in n:
        return 0

    # Ask user what they meant
    follow_up = input(f"I didn't understand '{reply}'. Does it mean yes? (yes/no): ").lower()

    # Add new word to correct column
    if follow_up in a: # affirmative
      pd.DataFrame([reply]).to_csv(affirmative, mode='a', header=False, index=False)
      return 1

    else:
      pd.DataFrame([reply]).to_csv(negative, mode='a', header=False, index=False)
      return 0

# MAIN PROGRAM

# Load score data
csv_path = "/content/drive/MyDrive/scoreData.csv"
df = pd.read_csv(csv_path)

# Make sure homework column exists
if 'homework' not in df.columns:
    df['homework'] = -1

# Separate features and target
X = df[['studied', 'slept', 'homework']].values
y = df['score'].values

# Train model
model = LinearRegression()
model.fit(X, y)

# User input
study = float(input("Enter hours studied: "))
sleep = float(input("Enter hours slept: "))
homework_input = input("Did you do homework well? (yes/no): ").lower()

answer = determine_answer(homework_input)

homework = 1 if answer == 1 else (0 if answer == 0 else -1)

predicted_score = model.predict([[study, sleep, homework]])[0]
predicted_score = max(0, min(predicted_score, 100))

print(f"My Predicted score for these inputs: {predicted_score:.1f}")

actual_score = float(input("What was your actual test score? "))

# Optional save
save = input("Do you want to save this data for training? (yes/no): ").lower()

if save == "yes":
    new_row = pd.DataFrame({
        "studied": [study],
        "slept": [sleep],
        "homework": [homework],
        "score": [actual_score]
    })
    df = pd.concat([df, new_row], ignore_index=True)
    df.to_csv(csv_path, index=False)

    X = df[['studied', 'slept', 'homework']].values
    y = df['score'].values
    model.fit(X, y)

# Make prediction
```
